```python
import matplotlib.pyplot as plt
import json

with open('ground_truth_2024-02-17_13-49-30.json') as f:
    data = json.load(f)

with open('tape2.obj') as f:
    tape = f.readlines()
```

```python
path = tape[4:664]
path = [x.split()[1:] for x in path]
path = [[float(x) for x in y] for y in path]
x_nb = [x[0] for x in path]
y_nb = [x[1] for x in path]
z_nb = [x[2] for x in path]
z_nb = [i * -1 for i in z_nb]
transform_x = -2.31
transform_z = 1.65
x_nb = [x + transform_x for x in x_nb]
z_nb = [x + transform_z for x in z_nb]
```

```python
x = [i['position']['x'] for i in data]
y = [i['position']['y'] for i in data]
z = [i['position']['z'] for i in data]
x = x[::50]
z = z[::50]
x = x[3:]
z = z[3:]
```

```python
# Figure size
plt.figure(figsize=(7, 6))
plt.axis('equal')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.grid()
plt.plot(x_nb, z_nb, c='red', linewidth=2)
plt.plot(x, z, '-o', c='tab:blue', markersize=10, fillstyle='none')
plt.legend(['Caminho de referncia', 'Trajetria do rob'], loc=(0.05, 0.2))
plt.show()
```
