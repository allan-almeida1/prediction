```python
import json
import matplotlib.pyplot as plt
import numpy as np

data = json.load(open('states_allan.json'))
data_tiago = json.load(open('states_tiago.json'))
print(len(data))
print(len(data_tiago))

de0 = [i['de0'] for i in data]
t = np.array([i['timestamp'] for i in data])
t = t - t[0]

de0_tiago = [i['de0'] for i in data_tiago]
t_tiago = np.array([i['timestamp'] for i in data_tiago])
t_tiago = t_tiago - t_tiago[0]

ltp = (0, 1)
len_allan = len(data)
len_tiago = len(data_tiago)


plt.plot(t_tiago[int(ltp[0] * len_tiago):int(ltp[1] * len_tiago)], de0_tiago[int(ltp[0] * len_tiago):int(ltp[1] * len_tiago)], 'r')
plt.plot(t[int(ltp[0] * len_allan):int(ltp[1] * len_allan)], de0[int(ltp[0] * len_allan):int(ltp[1] * len_allan)], 'b')
plt.ylim(-.27, .23)
plt.xlabel('Tempo (s)')
plt.ylabel('Z (m)')
plt.legend(['Método original', 'Método proposto'], loc='lower right')
plt.grid()
plt.show()

print(t[0] - t_tiago[0])
```
